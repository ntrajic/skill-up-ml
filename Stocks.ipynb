```python
# Applying sentiment analysis on real-world stock data

import pandas as pd

ft = pd.read_csv('stockerbot-export.csv', on_bad_lines='skip')
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
nltk.download('subjectivity')
nltk.download('movie_reviews')

sia = SentimentIntensityAnalyzer()
```

```python
ft.head()
```

```python
# find unuque stock ticker symbols above data contains
ft['symbols'].unique()
```

```python
# how many unique rows are there
ft['symbols'].value_counts()
```

```python
# analyze tweets: find only rows that satisfy condition to belong to one ticker related tweet
ft[ft['symbols'] == 'NFLX']
```

```python
# sort by timestamps
nt = ft[ft['symbols'] == 'NFLX']
nt_sorted = nt.sort_values(by='timestamp')
nt_sorted
```

```python
compound_scores = []

for index, row in nt_sorted.iterrows():
    tweet = row['text']
    analysis = sia.polarity_scores(tweet)
    compound_scores.append(analysis['compound'])
    
nt_sorted['sentiment_score'] = compound_scores # print
nt_sorted
```

```python
# plot, on x NFLX timestamp data, on y are compound scores:
import matplotlib.pyplot as plt

plt.plot(nt_sorted['timestamp'], compound_scores)
plt.ylabel('Compound Score')
plt.show()
```
